```python
#Load data from excel file
# 1. Load data from excel file
# 2. Clean data
# 3. Save data to csv file
import pandas as pd
import os

path = os.path.join("data", "Fredagsbarer_oversigt.xlsx")
df = pd.read_excel(path)
```

```python
import folium

# Create a Map centered around Aarhus
m = folium.Map(location=[56.1629, 10.2039], zoom_start=13)
```

```python
for index, row in df.iterrows():
    # check if the geo-location exists and is not NaN
    if pd.notna(row['Geo location']):
        try:
            # split the Geo location into latitude and longitude
            lat, lon = map(float, row['Geo location'].split(','))
            
            # format the popup string
            popup_text = f"""
                            <h4>{row['Name']}</h4><br>
                            <b>Address: </b>{row['Address']}<br>
                            <b>Opening hours: </b>{row['Opening hours']}<br>
                            <b>SoMe: </b><a href="{row['Social media']}">Link</a><br>
                            <b>Information: </b>{row['Information']}<br>
                            <b>Prices: </b>{row['Prices']}<br>
                        """
            # Create the folium marker and add to the map
            folium.Marker(
                location=[lat, lon],
                popup=folium.Popup(popup_text, max_width=250)
            ).add_to(m)
        except Exception as e:
            print(f"Error with row {index}: {e}")
```
